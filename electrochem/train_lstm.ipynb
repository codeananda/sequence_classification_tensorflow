```python
%load_ext autoreload
%autoreload 2

import wandb
from utils import train
import seaborn as sns

sns.set()
```

```python
for lr in [0.1, 0.01, 0.001, 1e-4, 1e-5, 1e-8, 1e-10]:
    num_runs = 4
    for i in range(num_runs):
        print(f'LR {lr} - RUN NUMBER {i+1}/{num_runs}')
        # Initalize wandb with project name
        run = wandb.init(project='electrochemistry',
                        config={
                            'data_processing': {
                                'a': 0,
                                'b': 1,
                                # Changing this will impact the batch_size!
                                'test_size': 0.2
                            },
                            'model': {
                                'model_type': 'LSTM',
                                'num_nodes': 64,
                                'input_shape': (1002, 1),
                                'loss': 'sparse_categorical_crossentropy',
                                'use_lr_scheduler': False,
                                'optimizer': 'adam',
                                'lr': lr,
                                'metrics': ['accuracy', 'sparse_top_k_categorical_accuracy']
                            },
                            'fit_model': {
                                'epochs': 100,
                                # Can only change this if you change test_size
                                'batch_size': 35,
                                'verbose': 2,
                                'shuffle': True
                            },
                            'callbacks': {
                                'patience': 10,
                                'restore_best_weights': True,
                                'baseline': None, # Set to None of there isn't one.
                                'average': 'micro'
                            },
                            'plotting': {
                                'start_plotting_epoch': 0
                            }  
                                })
        config = wandb.config # we use this to configure our experiment

        history = train(config)
        run.finish()
```
